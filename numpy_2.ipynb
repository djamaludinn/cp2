Накопленная оценка есть 0.3\*Проверочные + 0.4\*Домашние + 0.3\*Коллоквиум. Вычислите накопленные оценки и добавьте их в последний столбец матрицы (нужно добавить еще один новый столбец).

```python
# Веса следует брать из переменной weights, а не вписывать вручную
weights = np.array([0.3, 0.4, 0.3])
weights
```

```python
a = np.around(np.mean(marks[:, 0:3], axis = 1), decimals = 1)
a
```

```python
b = np.around(np.mean(marks[:, 3:6], axis = 1), decimals = 1)
b
```

```python
acc_mark = weights[0]*a+weights[1]*b+weights[2]*c
acc_mark = acc_mark[:, np.newaxis]
marks = np.hstack((marks, acc_mark))
marks
```
